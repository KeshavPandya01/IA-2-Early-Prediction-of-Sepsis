## Patient p008382

```python
df = pd.read_csv('Trainingset_A/p008382.psv',delimiter='|')
df
```

```python
# drop demographic and the outcome Variable

df = df.iloc[: , :-7]
df
```

## Visualizin Times series 

```python
plt.rcParams["figure.figsize"] = (12,30)
df.plot(subplots=True, legend=True)
plt.show()
```

## Box plot and outliers 

```python
dfb = pd.read_csv('Trainingset_A/p008382.psv',delimiter='|')
dfb = dfb.iloc[: , :-7]
plt.rcParams["figure.figsize"] = (14,8)
fig, axs = plt.subplots(3, 12)
fig.tight_layout()
for _,u in enumerate(dfb.columns):
    plt.subplot(3,12,_+1)
    dfb.boxplot(column=u)
    plt.title(u)
plt.show() 
```

## Histograms

```python
plt.rcParams["figure.figsize"] = (16,18)
fig, axs = plt.subplots(7, 5)
fig.tight_layout()
for _,u in enumerate(df.columns):
    plt.subplot(7,5,_+1)
    df[u].plot(kind='hist')
    plt.title(u)
plt.show()    
```

## Distributions

```python
import warnings
warnings.filterwarnings('ignore')
dfc = df.iloc[: , :34]

plt.rcParams["figure.figsize"] = (16,18)
fig, axs = plt.subplots(7, 5)
fig.tight_layout()
for _,u in enumerate(dfc.columns):
    try:
        plt.subplot(7,5,_+1)
        dfc[u].plot(kind='kde')
        plt.title(u)
    except:
        plt.title(u+'=>NaN')
```
